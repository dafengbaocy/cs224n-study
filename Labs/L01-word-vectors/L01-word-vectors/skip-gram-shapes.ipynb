# CS224N L01 — Skip-gram Data Flow: Center/Context Pairs & U/V Matrix Shapes

**Waypoint**: WP03 | **Anchor**: slides p25-p26, p28; notes sec 3.2 Eq.4-5

> **Why code?** Students mix up v_w (center), u_w (context), and matrix shapes.

```python
import numpy as np
corpus = ['banking crises into turning problems', 'turning banking into crises', 'problems into banking turning']
tokens = [s.lower().split() for s in corpus]
vocab = sorted(set(w for s in tokens for w in s))
w2i = {w:i for i,w in enumerate(vocab)}
i2w = {i:w for w,i in w2i.items()}
V, d = len(vocab), 4
print(f'|V|={V}, d={d}, vocab={vocab}')
```

## Skip-gram Pairs (window=2)

```python
pairs = []
for sent in tokens:
    for t,c in enumerate(sent):
        for j in range(-2,3):
            if j!=0 and 0<=t+j<len(sent):
                pairs.append((c, sent[t+j]))
print(f'Total pairs: {len(pairs)}')
for c,x in pairs[:8]: print(f'  {c} -> {x}')
print('  ...')
```

## U/V Matrices (notes sec 3.2)

```python
np.random.seed(42)
Vmat = np.random.randn(V,d).astype(np.float32)*0.5
Umat = np.random.randn(V,d).astype(np.float32)*0.5
print(f'V (center): {Vmat.shape}  U (context): {Umat.shape}')
for w in vocab[:3]:
    i=w2i[w]
    vstr = ', '.join(f'{x:+.4f}' for x in Vmat[i])
    ustr = ', '.join(f'{x:+.4f}' for x in Umat[i])
    print(f'  v_{w} = [{vstr}]')
    print(f'  u_{w} = [{ustr}]')
```

## Softmax P(o|c)

```python
c_word = 'banking'
v_c = Vmat[w2i[c_word]]
scores = Umat @ v_c
probs = np.exp(scores - scores.max()) / np.exp(scores - scores.max()).sum()
print(f'P(o|c="{c_word}"):')
for i in range(V):
    print(f'  {i2w[i]:>10s}  score={scores[i]:+.4f}  P={probs[i]:.4f}')
print(f'  sum = {probs.sum():.6f}')
```
